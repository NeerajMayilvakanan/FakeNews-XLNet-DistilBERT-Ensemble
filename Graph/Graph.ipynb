#Preparing CSV File

```python
# Create a dictionary to store the metrics data
metrics_data = {
    'Model': [],
    'Dataset': [],
    'Split': [],
    'Accuracy': [],
    'Precision' : [],
    'Recall' :[],
    'F1_Score' : []
}
```

```python
# Define the default values
model_names = ['XLNet', 'DistilBERT', 'Ensemble (Simple Average)', 'Ensemble (Majority Vote)']
datasets = ['Politifact', 'Gossipcop']
splits = ['Validation', 'Testing']
```

```python
for model in model_names:
    for dataset in datasets:
        for split in splits:
            print("Model : ",model)
            print("Dataset : ",dataset)
            print("Split : ",split)
            accuracy = input("Please enter accuracy: ")
            precision = input("Please enter precision: ")
            recall = input("Please enter recall: ")
            f1score = input("Please enter f1score: ")
            metrics_data['Model'].append(model)
            metrics_data['Dataset'].append(dataset)
            metrics_data['Split'].append(split)
            metrics_data['Accuracy'].append(accuracy)
            metrics_data['Precision'].append(precision)
            metrics_data['Recall'].append(recall)
            metrics_data['F1_Score'].append(f1score)
```

```python
# Create a DataFrame from the metrics data
metrics_df = pd.DataFrame(metrics_data)
```

```python
# Save the DataFrame to a CSV file
metrics_df.to_csv('metrics_data.csv', index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Define the models and ensembles
models = ['XLNet', 'DistilBERT', 'Ensemble (Simple Average)', 'Ensemble (Majority Vote)']
```

```python
# Define the metrics
metrics = ['Accuracy', 'Precision', 'Recall', 'F1_Score']
```

```python
for dataset in metrics_df['Dataset'].unique():
    for metric in metrics:
        plt.figure(figsize=(10, 6))
        sns.pointplot(x='Model', y=metric, data=metrics_df[metrics_df['Dataset'] == dataset], hue='Split', ci=None, dodge=True, palette='Set2')
        plt.title(f'{metric} Across Different Models and Ensembles for {dataset} Dataset')
        plt.xticks(rotation=45)
        plt.tight_layout()

        # Save the plot as an image (e.g., PNG)
        output_filename = f'{dataset}_{metric.replace(" ", "_").lower()}_pointplot.png'
        plt.savefig("/content/Group by metrics graph/"+output_filename, bbox_inches='tight')
        plt.show()
```

```python
y_values = range(70, 101, 5)
```

```python
# Reshape the DataFrame
melted_df = pd.melt(metrics_df, id_vars=['Model', 'Dataset', 'Split'], var_name='Metric', value_name='Value')

# Display the reshaped DataFrame
print(melted_df)
```

```python
for dataset in melted_df['Dataset'].unique():
    for split in melted_df['Split'].unique():
        # Filter the data for the current dataset and split
        dataset_split_data = melted_df[(melted_df['Dataset'] == dataset) & (melted_df['Split'] == split)]

        plt.figure(figsize=(12, 6))
        sns.set_style("whitegrid")
        sns.pointplot(x='Model', y='Value', data=dataset_split_data, hue='Metric', ci=None, dodge=True, palette='Set2')
        plt.title(f'{dataset} - {split} Metrics')
        plt.xticks(rotation=0)
        plt.yticks(y_values)
        plt.tight_layout()

        # Save the plot as an image (e.g., PNG)
        output_filename = f'{dataset}_{split.lower()}_metrics_pointplot.png'
        plt.savefig("/content/Hue_Metrics_for_DS_and_Split/"+output_filename, bbox_inches='tight')
        plt.show()
```
